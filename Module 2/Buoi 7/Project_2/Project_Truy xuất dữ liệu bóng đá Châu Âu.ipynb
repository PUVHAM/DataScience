```python
#Import libraries
import numpy as np
import pandas as pd
import sqlite3
from datetime import timedelta
import warnings
warnings.filterwarnings("ignore")
pd.set_option("display.max_columns", 100)
```

## Câu hỏi 1: Kết nối tới cơ sở dữ liệu và truy vấn thông tin của tất cả các bảng

Đọc https://docs.python.org/2/library/sqlite3.html

Học viên chỉ sử dụng câu lệnh SQL để truy vấn kế quả, không được sử dụng câu lệnh Pandas để tính toán ra kết quả.

```python
# Kết nối với cơ sở dữ liệu
# ĐIỀN CODE VÀO ĐÂY
conn = sqlite3.connect('./Data/database.sqlite')

# Lấy thông tin của tất cả bảng
tables = pd.read_sql("""SELECT *
                        FROM sqlite_master
                        WHERE type='table';""", conn)
tables
```

## Câu hỏi 2: Liệt kê các quốc gia có trong dữ liệu bảng "Country"

```python
# ĐIỀN CODE VÀO ĐÂY
result = pd.read_sql("SELECT * FROM Country", conn)
result
```

## Câu hỏi 3: Liệt kê các giải đấu trong bảng "League"

```python
# ĐIỀN CODE VÀO ĐÂY
result = pd.read_sql("SELECT * FROM League", conn)
result
```

## Câu hỏi 4: Liệt kê các giải đấu và thông tin quốc gia của từng league

Bảng League có quan hệ với bảng Country qua khóa country_id.

```python
# ĐIỀN CODE VÀO ĐÂY
result = pd.read_sql("""SELECT C.name AS Country, L.name AS League
                        FROM Country C 
                        INNER JOIN League L
                        ON C.id = L.country_id""", conn)
result
```

## Câu hỏi 5: Liệt kê các trận đấu trong bảng  “Match”
- Khi một đội là chủ nhà, đội đó là "home team". Đội còn lại là "away team"
- Trong bảng Match, mỗi dòng dư liệu là một trận đấu của một đội chủ nhà và đội khách, tương ứng với kết quả ghi bàn đội nhà (home goal) và đội khách (away goal).

```python
# ĐIỀN CODE VÀO ĐÂY
result = pd.read_sql("SELECT * FROM Match", conn)
result
```

## Câu hỏi 6: Liệt kê các trận đấu và thông tin về league và country tương ứng

Bảng Match có quan hệ với :
 - Bảng Country thông qua khóa country_id
 - Bảng League thông qua khóa League_id

Sử dụng câu lệnh join để kết nối dư liệu từ 3 bảng trên.


```python
# ĐIỀN CODE VÀO ĐÂY
result = pd.read_sql("""SELECT *
                        FROM Match M
                        JOIN League L
                        ON M.League_id = L.id
                        JOIN Country C
                        ON M.country_id = C.id""", conn)
result
```

## Câu hỏi 7: Liệt kê số lượng trận đấu của mỗi giải đấu bao gồm cả tên giải đấu, sắp xếp theo thứ tự giảm dần số trận đấu

```python
# ĐIỀN CODE VÀO ĐÂY
result = pd.read_sql("""SELECT L.name AS League, COUNT(*) MatchCount
                        FROM League L
                        INNER JOIN Match M
                        ON M.League_id = L.id
                        GROUP BY L.name""", conn)
result
```

## Câu hỏi 8: Liệt kê tổng số bàn thắng của đội nhà và đội khách trong mỗi giải đấu
- Sử dụng câu lệnh Group By

```python
# ĐIỀN CODE VÀO ĐÂY
result = pd.read_sql("""SELECT L.name as League, SUM(home_team_goal) AS TotalHomeTeamGoal, SUM(away_team_goal) AS TotalAwayTeamGoal
                        FROM Match M
                        INNER JOIN League L
                        ON M.League_id = L.id
                        GROUP BY L.name""", conn)
result
```

## Câu hỏi 9: Liệt kê thông tin các đội từ bảng “Team”

```python
# ĐIỀN CODE VÀO ĐÂY
result = pd.read_sql("SELECT * FROM Team", conn)
result
```

## Câu hỏi 10: Liệt kê 20 đội với số bàn thắng sân nhà cao nhất
- Sử dung câu lênh GROUP BY, LIMIT

```python
# ĐIỀN CODE VÀO ĐÂY
result = pd.read_sql("""SELECT T.team_long_name as Team, SUM(home_team_goal) AS TotalHomeTeamGoal
                        FROM Match M
                        INNER JOIN Team T
                        ON M.home_team_api_id = T.team_api_id
                        GROUP BY T.team_long_name
                        ORDER BY TotalHomeTeamGoal DESC
                        LIMIT 20""", conn)
result
```

## Câu hỏi 11: Liệt kê 20 đội có số bàn thắng trên sân khách cao nhất
- Sử dụng câu lênh GROUP BY, LIMIT

```python
# ĐIỀN CODE VÀO ĐÂY
result = pd.read_sql("""SELECT T.team_long_name as Team, SUM(away_team_goal) AS TotalAwayTeamGoal
                        FROM Match M
                        INNER JOIN Team T
                        ON M.away_team_api_id = T.team_api_id
                        GROUP BY T.team_long_name
                        ORDER BY TotalAwayTeamGoal DESC
                        LIMIT 20""", conn)
result
```

## Câu hỏi 12: Liệt kê tên các đội bóng và tổng số bàn thắng mỗi đội, sắp xếp  theo số lượng giảm dần

- Tổng số bàn thắng của một đội bao gồm cả số bàn thắng ở sân nhà và sân khách.
- Sử dụng câu lệnh Union  

```python
# ĐIỀN CODE VÀO ĐÂY
result = pd.read_sql("""SELECT Team, SUM(TotalGoals) AS TotalGoals
                        FROM (
                            SELECT T.team_long_name AS Team, SUM(M.away_team_goal) AS TotalGoals
                            FROM Match M
                            INNER JOIN Team T 
                            ON M.away_team_api_id = T.team_api_id
                            GROUP BY T.team_long_name
                            
                            UNION ALL
                            
                            SELECT T.team_long_name AS Team, SUM(M.home_team_goal) AS TotalGoals
                            FROM Match M
                            INNER JOIN Team T 
                            ON M.home_team_api_id = T.team_api_id
                            GROUP BY T.team_long_name
                        ) AS CombinedGoals
                        GROUP BY Team
                        ORDER BY TotalGoals DESC""", conn)

result
```

## Câu hỏi 13: Liệt kê tên các đội bóng (long name) và tổng số trận đấu đội đó tham gia, sắp xếp theo thứ tự giảm dần.

```python
# ĐIỀN CODE VÀO ĐÂY
result = pd.read_sql("""SELECT Team, COUNT(*) AS TotalMatches
                        FROM (
                            SELECT T.team_long_name AS Team
                            FROM Match M
                            INNER JOIN Team T 
                            ON M.away_team_api_id = T.team_api_id
                            
                            UNION ALL
                            
                            SELECT T.team_long_name AS Team
                            FROM Match M
                            INNER JOIN Team T 
                            ON M.home_team_api_id = T.team_api_id
                        ) AS CombinedMatches
                        GROUP BY Team
                        ORDER BY TotalMatches DESC""", conn)

result
```

## Câu hỏi 14: Liệt kê số trận thắng, thua và hòa của mỗi đội.

- Nếu một đội có số lượng bàn thắng trên sân nhà > số lượng bàn thắng trên sân khách của chính đội đó, thì đội đó sẽ "thắng" trong trận đấu này
- Nếu một đội có số lượng bàn thắng trên sân nhà < số lượng bàn thắng trên sân khách của chính đội đó, thì đội đó sẽ "thua" trong trận đấu này
- Nếu một đội có số lượng bàn thắng trên sân nhà = số lượng bàn thắng trên sân khách của chính đội đó, thì đội đó sẽ "hòa" trong trận đấu này
- Sử dụng câu lệnh "Case When"


```python
result = pd.read_sql("""
                        SELECT Team, SUM(HomeGoals) AS TotalHomeGoals, SUM(AwayGoals) AS TotalAwayGoals,
                               CASE
                                   WHEN SUM(HomeGoals) > SUM(AwayGoals) THEN 'thắng'
                                   WHEN SUM(HomeGoals) < SUM(AwayGoals) THEN 'thua'
                                   ELSE 'hoà'
                               END AS Trận
                        FROM (
                            SELECT 
                                T.team_long_name AS Team,
                                M.home_team_goal AS HomeGoals,
                                M.away_team_goal AS AwayGoals
                            FROM Match M
                            INNER JOIN Team T ON M.home_team_api_id = T.team_api_id
                            
                            UNION ALL
                            
                            SELECT 
                                T.team_long_name AS Team,
                                M.away_team_goal AS HomeGoals,
                                M.home_team_goal AS AwayGoals
                            FROM Match M
                            INNER JOIN Team T ON M.away_team_api_id = T.team_api_id
                        ) AS CombinedMatches
                        GROUP BY Team
                        ORDER BY Team
                    """, conn)

result
```

## Câu hỏi 15: Liệt kê 10 đội có số trận thắng nhiều nhất

```python
result = pd.read_sql("""
                        SELECT Team,
                               SUM(CASE WHEN Result = 'thắng' THEN 1 ELSE 0 END) AS Wins
                        FROM (
                            SELECT 
                                T.team_long_name AS Team,
                                CASE 
                                    WHEN M.home_team_goal > M.away_team_goal THEN 'thắng'
                                    WHEN M.home_team_goal < M.away_team_goal THEN 'thua'
                                    ELSE 'hoà'
                                END AS Result
                            FROM Match M
                            INNER JOIN Team T ON M.home_team_api_id = T.team_api_id
                            
                            UNION ALL
                            
                            SELECT 
                                T.team_long_name AS Team,
                                CASE 
                                    WHEN M.away_team_goal > M.home_team_goal THEN 'thắng'
                                    WHEN M.away_team_goal < M.home_team_goal THEN 'thua'
                                    ELSE 'hoà'
                                END AS Result
                            FROM Match M
                            INNER JOIN Team T ON M.away_team_api_id = T.team_api_id
                        ) AS AllMatches
                        GROUP BY Team
                        ORDER BY Wins DESC
                        LIMIT 10; """, conn)
result
```

## Câu hỏi 16: Liệt kê tỉ lệ phần trăm về số trận của mỗi giải trên tổng tất cả các trận

```python
# ĐIỀN CODE VÀO ĐÂY
result = pd.read_sql("""
                        SELECT Team,
                               SUM(CASE WHEN Result = 'thắng' THEN 1 ELSE 0 END) AS Wins,
                               COUNT(*) AS TotalMatches,
                               CAST(SUM(CASE WHEN Result = 'thắng' THEN 1 ELSE 0 END) AS FLOAT) / COUNT(*) AS WinRate
                        FROM (
                            SELECT 
                                T.team_long_name AS Team,
                                CASE 
                                    WHEN M.home_team_goal > M.away_team_goal THEN 'thắng'
                                    WHEN M.home_team_goal < M.away_team_goal THEN 'thua'
                                    ELSE 'hoà'
                                END AS Result
                            FROM Match M
                            INNER JOIN Team T ON M.home_team_api_id = T.team_api_id
                            
                            UNION ALL
                            
                            SELECT 
                                T.team_long_name AS Team,
                                CASE 
                                    WHEN M.away_team_goal > M.home_team_goal THEN 'thắng'
                                    WHEN M.away_team_goal < M.home_team_goal THEN 'thua'
                                    ELSE 'hoà'
                                END AS Result
                            FROM Match M
                            INNER JOIN Team T ON M.away_team_api_id = T.team_api_id
                        ) AS AllMatches
                        GROUP BY Team
                        ORDER BY WinRate DESC
                        LIMIT 10; """, conn)
result
```

## Câu hỏi 17: Liệt kê tỉ lệ phần trăm về số bàn ghi được của từng giải so với tổng số bàn của tất cả các trận.

```python
# ĐIỀN CODE VÀO ĐÂY
result = pd.read_sql("""
                        SELECT
                            M.league_id,
                            L.name AS league_name,
                            COUNT(M.id) AS total_goals,
                            (COUNT(M.id) * 100.0 / (SELECT COUNT(*) FROM Match)) AS goal_percentage
                        FROM
                            Match M
                        INNER JOIN
                            League L ON M.league_id = L.id
                        GROUP BY
                            M.league_id, league_name
                        ORDER BY
                            goal_percentage DESC;
                            """, conn)
result
```

## Câu hỏi 18: Liệt kê tổng số bàn ghi được của từng giải đấu theo từng mùa giải.

```python
# ĐIỀN CODE VÀO ĐÂY
result = pd.read_sql("""
                        SELECT
                            M.league_id,
                            L.name AS league_name,
                            COUNT(M.id) AS total_goals,
                            Season
                        FROM
                            Match M
                        INNER JOIN
                            League L ON M.league_id = L.id
                        GROUP BY
                            league_name, Season
                        ORDER BY
                            total_goals DESC;
                            """, conn)
result
```
